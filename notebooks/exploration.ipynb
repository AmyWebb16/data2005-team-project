```python
import os

current = os.getcwd()

while True:
    items = os.listdir(current)
    if "src" in items and "data" in items:
        os.chdir(current)
        break

    parent = os.path.dirname(current)
    if parent == current:
        raise RuntimeError("Could not find project root.")
    current = parent

print("Project root:", os.getcwd())
```
